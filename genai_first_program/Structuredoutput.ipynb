### Model Initialisation

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model=init_chat_model("groq:llama-3.1-8b-instant")
model.invoke("Hello, how are you?")
```

### Pydentic - Typedict - Dataclasses

Typeddict-> Without data validation in dictionary format
Dataclasses-> Without dictionary format


### Pydentic
provides field validation,descriptions and nested structures

```python
from pydantic import BaseModel,Field

class movie(BaseModel):
    title:str=Field(description="The title of the Movie")
    year:int=Field(description="The year of the Movie was released")
    rating:float=Field(description="The rating of the Movie")
    hit:bool=Field(description="The movie was hit or flopped")
    
```

```python
model_structured_output=model.with_structured_output(movie)
model_structured_output
```

```python
response=model_structured_output.invoke("Tell me about Titanic movie?")
print(response)
```

```python
## Message output along with parsed structure

model_structured_output=model.with_structured_output(movie,include_raw=True)
response=model_structured_output.invoke("Tell me about Titanic movie?")
print(response)
```

```python
## Nested structure

from pydantic import BaseModel,Field

class Actor(BaseModel):
    name:str
    role:str

class Movie(BaseModel):
    title:str
    year:int
    cast:list[Actor]
    geners:list[str]

model_structured_output=model.with_structured_output(Movie)
response=model_structured_output.invoke("Tell me about Titanic movie?")
print(response)

```
