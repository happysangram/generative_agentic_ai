#### Streaming and Batch Processing and Invoke

```python
response = llm.batch(["Hello, how are you?", "what is the capital of France?", "what is the currency of France?"])

for res in response:
    print(res.content, end="", flush=True)
```

```python
for chunk in llm.stream("Write a essay with 100 words"):
    print(chunk.content, end="", flush=True)
```
